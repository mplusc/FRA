```python
# Import Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reference the file where the CSV is located
data_csv_path = 'resources/data.csv'

# Import the data into a Pandas DataFrame
data_df = pd.read_csv(data_csv_path)
data_df.head()
```

```python
# Get a reference to the column names
data_df.columns

# Renaming the column headers
data_df = data_df.rename(columns={'Autos':'Auto Sales (thousands)',
                                   'Var01':'US Real GDP Growth',
                                   'Var02':'US Nominal GDP Growth',
                                   'Var03':'US Real Disposable Income Growth',
                                   'Var04':'US Nominal Disposable Income Growth',
                                   'Var05':'US Unemployment Rate'})
```

```python
# Removing the blanks from column headers
data_df = data_df.dropna(axis=1,how='all')

# Removing the blanks from rows
data_df = data_df[np.isfinite(data_df['Time_Period'])]

data_df.tail(10)
```

```python
#Cleaning data for model forecasting with machine learning
data_df_model = data_df[np.isfinite(data_df['Auto Sales (thousands)'])]
```

```python
#Preview data for model
data_df_model.tail()
```

```python
#export dataframe to csv for model

data_df_model.to_csv('resources/model_data.csv', index=False)
```

```python
#export dataframe to csv for model

data_df.to_csv('resources/full_data.csv', index=False)
data_df.tail(10)
```
